```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de densidad
def f(x):
    if -2 <= x <= 2:
        return (90*(x**8))*(1-x)
    else:
        return 0

# Crear un rango de valores de x
x = np.linspace(0, 1, 1000)  # 1000 puntos entre 3 y 5

# Calcular los valores de la función de densidad para cada valor de x
y = [f(i) for i in x]

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = 90x^8 (1-x)')
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Densidad de probabilidad f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def equation(x):
    return 9*x**10 - 10*x**9 + 0.75

def bisection_method(func, a, b, tolerance):
    if func(a) * func(b) >= 0:
        return None
    while (b - a) / 2 > tolerance:
        c = (a + b) / 2
        if func(c) == 0:
            return c
        elif func(c) * func(a) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

root = bisection_method(equation, 0, 1, 0.0001)
print("El percentil 75 de la distribución es:", root)
```
